```python
arr = np.array([0,1,2,3,4,5])
print(arr)
arr1 = np.array([2,2,2,2,2,2])
print(arr1)

#to increase size of the graph
plt.figure(figsize = (8, 5.75)) #figsize parameter of plt.figure to specify the size of graph

#to plot two arrays in the same graph, use plt twice
plt.plot(arr, linewidth = 5, label = "constant increase") #linewidth parameter is used to specify that line's width
plt.plot(arr1, color = 'red', linestyle = ':', label = "constant value") 
#color parameter is used to specify the line's color
#linestyle parameter can change the type of line
    
#giving the graph a title
plt.title("Plotting Arrays with Line Graphs", loc = 'center', y = 1.02, fontsize = 15, fontweight = 'heavy')
#using the location parameter to adjust title's location
#y parameter ranges from 0 - 1, 0 being the bottom and 1 being above the graph boundary
#fontsize parameter adjusts the size of title
#fontweight parameter - normal, bold, heavy, light

plt.legend(title = "Line Representation", loc = "upper left") #shows the labels for lines on the graph

plt.xlabel("Independent Variable", fontsize = 15, fontweight = 'bold') #gives the x-axis a label
plt.ylabel("Dependent Variable", fontsize = 15) #gives the y-axis a label

plt.xticks(range(0,len(arr)),('jan','feb','mar','apr','may','jun'))
#this function has two parameters, the first one is the range of what you want to change
#the second one is what you want to change it to

plt.show() #displays the graph
```

## 3. Scatter Plots using matplotlib and numpy

```python
#creating data to plot

arr = np.arange(30)

arr1 = np.random.randint(0,30,30)
#first parameter is starting of range
#second parameter is ending of range
#third parameter is the number of randoms you want

#trying to manipulate arr1
arr1 = arr + np.random.randint(0,30,30)

size = np.random.randint(0,100,30) 
#the values in this array 

plt.scatter(arr,arr1, c = 'green', s = size) #generates a scatterplot 
#first parameter is the values on x axis
#second parameter is the values on y axis
#c stands for color
#s stands for size

plt.title("Randomly Generated Scatter Plot")
plt.ylabel("Random Values")
plt.xlabel("Specified Values")

plt.show()
```

## 9. Forming graphs using HealPy

### Examples

```python
from astropy.coordinates import SkyCoord
import healpy as hp
import numpy as np

#function to convert data to a healpix sky map
def cat2hpx(lon, lat, nside, radec=True):
    npix = hp.nside2npix(nside)

    if radec:
        eq = SkyCoord(lon, lat, 'icrs', unit='deg')
        l, b = eq.galactic.l.value, eq.galactic.b.value
    else:
        l, b = lon, lat

    # conver to theta, phi
    theta = np.radians(90. - b)
    phi = np.radians(l)

    # convert to HEALPix indices
    indices = hp.ang2pix(nside, theta, phi)

    idx, counts = np.unique(indices, return_counts=True)

    # fill the fullsky map
    hpx_map = np.zeros(npix, dtype=int)
    hpx_map[idx] = counts

    return hpx_map
```

```python
l = np.random.uniform(-180, 180, 20000)
b = np.random.uniform(-90, 90, 20000)

hpx_map = cat2hpx(l, b, nside=32, radec=False)

hp.mollview(np.log10(hpx_map + 1))  #you use log scale to plot
```

```python
hp.orthview(np.log10(hpx_map+1), rot=[0, 90])
hp.graticule(color='white')
```
